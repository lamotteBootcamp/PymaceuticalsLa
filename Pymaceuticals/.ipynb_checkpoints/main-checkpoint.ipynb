```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

# Display the data table for preview
mouse_and_study_df = pd.merge(study_results,mouse_metadata, how ="left", on = ["Mouse ID","Mouse ID"])
#mouse_and_study_df = pd.DataFrame(data = mouse_and_study)
mouse_and_study_df.head()

#DONE
```

```python
# Checking the number of mice.
mouse_count = len(mouse_and_study_df["Mouse ID"].unique())
mouse_count
#DONE
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse = mouse_and_study_df.loc[mouse_and_study_df.duplicated(subset=['Mouse ID','Timepoint',]),'Mouse ID'].unique()
duplicate_mouse
#duplicate_mouse = mouse_and_study_df[mouse_and_study_df.duplicated(keep=False)]
#duplicate_mouse
#DONE
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_table=mouse_and_study_df.loc[mouse_and_study_df["Mouse ID"]=="g989", :]
duplicate_mouse_table
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#clean_mouse_and_study_df = mouse_and_study_df.drop_duplicates(keep=False,inplace=True)
#clean_mouse_and_study_df.head()
clean_mouse_and_study_df=mouse_and_study_df[mouse_and_study_df['Mouse ID'].isin(duplicate_mouse)==False]
clean_mouse_and_study_df
#DONE
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_count=len(clean_mouse_and_study_df['Mouse ID'].unique())
clean_mouse_count
#DONE
```

```python
###Summary Statistics
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mouse_treat_summary = clean_mouse_and_study_df.groupby('Drug Regimen')

mouse_treat_mean = mouse_treat_summary["Tumor Volume (mm3)"].mean()

mouse_treat_median = mouse_treat_summary["Tumor Volume (mm3)"].median()

mouse_treat_variance = mouse_treat_summary["Tumor Volume (mm3)"].var()

mouse_treat_standdev = mouse_treat_summary["Tumor Volume (mm3)"].std()

mouse_treat_sem = mouse_treat_summary["Tumor Volume (mm3)"].sem()

#DONE
```

```python
mouse_treat_summary_table = pd.DataFrame({
                            "Mean Tumor Volume":mouse_treat_mean,
                            "Median Tumor Volume":mouse_treat_median,       
                            "Tumor Volume Variance":mouse_treat_variance,
                            "Tumor Volume Std. Dev.":mouse_treat_standdev,
                            "Tumor Volume Std. Error":mouse_treat_sem,
})
mouse_treat_summary_table

#DONE
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.

mouse_aggregate = clean_mouse_and_study_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean','median','var','std','sem'])
mouse_aggregate
#aggregating_data = clean_mouse_and_study_df.groupby('Drug Regimen')
#mouse_aggregate = aggregating_data.agg(['mean','median','var','std','sem'])["Tumor Volume (mm3)"]
#mouse_aggregate
#DONE
```

```python
###Quartiles, Outliers and Boxplots
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_tumor = clean_mouse_and_study_df.groupby(["Mouse ID"])["Timepoint"].max()
max_tumor=max_tumor.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data = max_tumor.merge(clean_mouse_and_study_df,on=['Mouse ID','Timepoint'],how="left")
merged_data
#DONE
```

```python
# Put treatments into a list for for loop (and later for plot labels)
four_treat_list = ['Capomulin','Ramicane','Infubinol','Ceftamin']

counter = 0

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list =[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in four_treat_list:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_vol = merged_data.loc[merged_data["Drug Regimen"]==drug,'Tumor Volume (mm3)']
    
    # add subset 
    tumor_vol_list.append(final_tumor_vol)   
    
     Determine outliers using upper and lower bounds
    quartiles = tumor_vol_list.quantile([0.25,0.5,0.75])
    lowerq=quartiles[0.25]
    upperq=quartiles[0.75]
    iqr=upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    if counter ==0:
    
        print(f"{drug}'s potential outliers: {lower_bound},{upper_bound}")
    
    counter += 1
print(tumor_vol_list)
    
    #print(f"Ramicane's potential outliers:")
    #print(f"Ramicane's potential outliers:")
    #print(f"Ramicane's potential outliers:")
    
```
